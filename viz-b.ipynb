```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# car_crashes
data = sns.load_dataset('car_crashes')
# data.head(10)
df = pd.DataFrame(data)
df.head(30)
```

```python
# car_crashes = sns.load_dataset("car_crashes")
# sns.relplot(data=car_crashes, x="speeding", y="total", multiple="stack")
g = sns.relplot(data=df, x="speeding", y="total")
g.ax.axline(xy1=(10, 2), slope=.2, color="b", dashes=(5, 2))
```

#### Drawing the relation between the total accedint vs the speeding : 
- the more a person speed he is more likley to hav an accedint
